**6. What does the `na.rm` argument do in `geom_point()`? What is the default value of the argument? Create a scatterplot where you successfully use this argument set to `TRUE`.**

_**Answer:**_ `na.rm` sets the warning about missing values on and off. Its default is `FALSE`, meaning it will warn against missing values. When set to true, the functions won't warn anymore.

Note, however, that `na.rm` must be set to `TRUE` in all subsequent function which deal with the points themselves as well, like `geom_smooth()` in this case 👇

```r
ggplot(
    data = penguins,
    mapping = aes(x = bill_length_mm, y = bill_depth_mm, color = species)
) +
    geom_point(mapping = aes(shape = species), na.rm = TRUE) +
    geom_smooth(method = "lm", na.rm = TRUE) +
    labs(
        title = "Bill depth vs Bill length",
        subtitle = "Trends for Adelie, Chinstrap, and Gentoo Penguins",
        x = "Bill length (mm)", y = "Bill depth (mm)",
        color = "Species", shape = "Species"
    ) +
    scale_color_colorblind()
```

**10. Will these two graphs look different? Why/why not?**

_**Answer:**_ No. `data` and `mapping` can be set both at the global level(at `ggplot()`) or at the local level(in e.g. `geom_smooth()`). But setting these parameters only once at the global level avoids repetition if they're gonna be used thoughout the plot.

```r
ggplot(
    data = penguins,
    mapping = aes(x = flipper_length_mm, y = body_mass_g)
) +
    geom_point() +
    geom_smooth()
```

```r
ggplot() +
    geom_point(
        data = penguins,
        mapping = aes(x = flipper_length_mm, y = body_mass_g)
    ) +
    geom_smooth(
        data = penguins,
        mapping = aes(x = flipper_length_mm, y = body_mass_g)
    )
```

## Plotting categorical values

When plotting categorical values, R will not plot them in a particular order.

```r
ggplot(penguins, aes(x = species)) +
    geom_bar()
```

If we want to order the bars - e.g., by frequency - we need to convert the values to `factors` explicitly and reorder them.

```r
class(penguins$species)
```

```r
ggplot(penguins, aes(x = fct_infreq(species))) +
    geom_bar()
```

## Visualizing relationships

### A numerical and a categorical variable

```r
ggplot(penguins, aes(x = species, y = body_mass_g)) +
    geom_boxplot()
```

```r
ggplot(penguins, aes(x = body_mass_g, color = species)) +
    geom_density() # nolint: indentation_linter.
```
